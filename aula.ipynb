```python
df = pd.read_csv("dados/FRA3-FRA6_cleaned_feature_engineered.csv.zip")
df_aux = df.copy()
df
```

```python

def create_scaled_dataset(X, y):
    from sklearn.model_selection import cross_val_score, train_test_split
    from sklearn.preprocessing import StandardScaler
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=0)

    std_scaler = StandardScaler()
    X_scaled_train = pd.DataFrame(std_scaler.fit_transform(
        X_train), columns=X_train.columns, index=X_train.index)
    X_scaled_test = pd.DataFrame(std_scaler.transform(
        X_test), columns=X_train.columns, index=X_test.index)
    return X_scaled_train, X_scaled_test, y_train, y_test

def create_scaled_dataset_for_cross(X,y):
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import StandardScaler
    from sklearn.utils import shuffle

    X, y = shuffle(X, y)

    std_scaler = StandardScaler()
    X = pd.DataFrame(std_scaler.fit_transform(X),columns=X.columns,index=X.index)
    
    return X,y
```

## Criando dados de treino

```python
target = "Vert_irreg_right_rail"

y = df[target]
X = df.drop(columns=target)


X_train, X_test, y_train, y_test = create_scaled_dataset(X, y)
```

### Definindo os modelos a serem otimizados

```python
reg = RegModels(modo="load", X_trainS=X_train, X_testS=X_test, y_trainS=y_train, y_testS=y_test)

path = "fitted_models_copy"
minhaLista = {'Lasso': 20,
              'Random_Forest':  5,
              'Elastic_Net':  5,
              'Ada_Boost':  4,
              'Ridge':  20,
              'XGBR': 8,
              'Extra_Trees': 10,
              'Cat_Boost': 4,
              'Light_Boost': 4,
              'KNN_Regressor': 5,
              'SGD_Reg': 6
              }

minhaLista ={'Lasso': 4,'Ridge':  20}

reg.fit_models(minhaLista,path)
```

```python


reg.models_performace(['Lasso','Ridge','KNN_Regressor'])
```

```python
from sklearn.ensemble import StackingRegressor

from sklearn import linear_model 
estimators = [('Lasso',linear_model.Lasso()),('ridge',linear_model.Ridge())] 

stacking_regressor = StackingRegressor(
            estimators=estimators, final_estimator= linear_model.RidgeCV())
stacking_regressor.fit(X_train, y_train)
stacking_regressor.score(X_test,y_test)
```

## Bagging

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import linear_model

model = linear_model.Lasso()

beg = BaggingRegressor(base_estimator=model,n_estimators=50,n_jobs=2,oob_score=True)
beg.fit(X_train,y_train)
```

```python
beg.score(X_test,y_test)
```

## Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score

X_cross,y_cross = create_scaled_dataset_for_cross(X,y)


kfold = KFold(n_splits=10, random_state=None, shuffle=False)

rmseList = cross_val_score(beg, X_cross, y_cross, cv=kfold, n_jobs=2, scoring="neg_root_mean_squared_error")
```
